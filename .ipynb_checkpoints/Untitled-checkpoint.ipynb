```python
import numpy as np
import pandas as pd
import json
import datetime
import matplotlib.pyplot as plt

from sqlalchemy import create_engine
from pandas.io import sql

import QSTK.qstkutil.qsdateutil as du
import QSTK.qstkutil.tsutil as tsu
import QSTK.qstkutil.DataAccess as da

%matplotlib inline
engine = create_engine('mysql://root:testdb@127.0.0.1/tstest?charset=utf8')
```

```python
#get econ indicator
cnx = engine.raw_connection() # option-2
econ = sql.read_sql("SELECT * FROM tstest.macro_detail where indicID = 'M010000003'", cnx)
cnx.close()

#convert date format and set publish date as index
econ['publishTime'] = pd.to_datetime(econ['publishTime']).apply(lambda x: x.date())
econ = econ.set_index('publishTime')
```

```python
#create start date, end date and symbol list
cnx = engine.raw_connection() # option-2
eqs = sql.read_sql("SELECT * FROM equ_info_a", cnx)
cnx.close()
sd = datetime.date.today() - datetime.timedelta(days=150)
ed = datetime.date.today()
symbols = eqs.secID

#function to get data from database
def get_bars(ids,sd,ed):
    ids = tuple(ids.apply(str))
    cnx = engine.raw_connection() # option-2
    eodp = pd.DataFrame()
    #create query pull price history for all stocks
    query = "SELECT secID,tradeDate, closePrice  FROM mkt_price where secID in %r and tradeDate between %r and %r" %(ids, str(sd),str(ed))
    edop = sql.read_sql(query, cnx)    
        
    return edop

#reshape data
data = get_bars(symbols.head(),sd,ed)
bars = data.pivot('tradeDate','secID')['closePrice']
```

```python
ldt_timestamps = bars.index
d_data = bars['000001.XSHE'].values
```

```python
na_price = bars.values
na_normalized_price = na_price / na_price[0, :]
plt.clf()
plt.plot(ldt_timestamps, na_normalized_price)
plt.legend(symbols)
plt.ylabel('Close')
plt.xlabel('Date')
plt.show()
```
