```python
import pandas as pd
import numpy as np
import talib
import statsmodels
import matplotlib.pyplot as plt

from sqlalchemy import create_engine
from pandas.io import sql

%matplotlib inline
engine = create_engine('mysql://root:testdb@127.0.0.1/tstest?charset=utf8')
```

```python
cnx = engine.raw_connection() # option-2
xx = sql.read_sql("SELECT tradeDate, closePrice  FROM mkt_price where secID='000001.XSHE'", cnx)
cnx.close()
```

```python
close = xx[-500:].ix[:,1]
```

```python
avg = talib.SMA(close.values, timeperiod=30)
macd, macdsignal, macdhist = talib.MACD(close.values, fastperiod=12, slowperiod=26, signalperiod=9)
```

```python
type(close.values)
```

```python
plt.clf()
fig = plt.figure()
plt.subplot(2, 1, 1)
plt.plot(xx[-500:].ix[:,0], zip(close,avg))
plt.subplot(2, 1, 2)
plt.plot(xx[-500:].ix[:,0], zip(macd, macdsignal, macdhist))
fig.autofmt_xdate(rotation=45)
```

```python
data =  pd.DataFrame({'date': xx[-500:].ix[:,0],'close':xx[-500:].ix[:,1],'macd':macd, 'macdsignal':macdsignal, 'macdhist':macdhist,'avg':avg})
```

```python
data = data.set_index('date')
```

```python
plt.clf()

data.plot(y=['close','avg'],title='moving avg')

data.plot(y=['macd','macdsignal','macdhist'],title='MACD')
fig.autofmt_xdate(rotation=45)
```
